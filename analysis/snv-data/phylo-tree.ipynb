# Load and show tree

```python
from Bio import Phylo
import matplotlib
import matplotlib.pyplot as plt

tree = Phylo.read('align_hf.phy_phyml_tree.txt', 'newick')
tree.ladderize()

fig = plt.figure(figsize=(8, 8), dpi=100)
axes = fig.add_subplot(1, 1, 1)
Phylo.draw(tree, axes=axes)
plt.show()
fig.savefig('phylo-tree-hf.pdf', dpi=100)
```
